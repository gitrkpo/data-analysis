```python
import pandas as pd
from sqlalchemy import create_engine
import schedule
import time
import logging

# Setup logging
logging.basicConfig(filename='etl_process.log', level=logging.INFO)

# SQL Server connection details
server = "DESKTOP-KFV9TLF\\MSSQLSERVEROO"
database = "TEST"

# Create SQLAlchemy engine using Windows Authentication
connection_string = f"mssql+pyodbc://@{server}/{database}?trusted_connection=yes&driver=ODBC+Driver+17+for+SQL+Server"
engine = create_engine(connection_string)

def etl_process():
    try:
        # Step 1: Extract - Load new data from CSV
        logging.info("Starting extraction process...")
        data = pd.read_csv('dirtydata.csv')  # Read CSV file
        logging.info(f"Extracted {len(data)} rows.")

        # Step 2: Transform - Clean data
        logging.info("Starting data transformation process...")
        data['Calories'] = data['Calories'].fillna(data['Calories'].mean())  # Fill NaN in Calories with the mean value
        data['Calories'].fillna(4, inplace=True)  # Replace any remaining NaNs in Calories with 4

        data['Date'] = data['Date'].dropna()  # Drop NaN in Date
        data.dropna(subset=['Date'], inplace=True)  # Drop rows with NaN in Date

        data.loc[data.Duration > 100, 'Duration'] = 75  # Convert anything more than 100 in Duration to 75
        data['Date'] = pd.to_datetime(data['Date'])  # Convert Date to datetime format
        data['Duration'] = data['Duration'].replace(450, 58)  # Replace 450 with 58 in Duration

        # Filter: Pulse > 100
        filtered_data = data[data['Pulse'] > 100]
        logging.info(f"Filtered data to {len(filtered_data)} rows.")

        # Step 3: Load - Insert into SQL Server
        logging.info("Starting data load process...")
        filtered_data[['Duration', 'Date', 'Pulse']].to_sql("new_sales_data", engine, if_exists="replace", index=False)
        logging.info("Data loaded successfully into SQL Server.")

    except Exception as e:
        logging.error(f"ETL process failed: {e}")
        raise  # Reraise the exception after logging

    logging.info("ETL process completed successfully!")  # Log successful completion
    print("ETL process completed!")

# Schedule the job to run weekly, for example, every Monday at 8:00 AM
schedule.every().monday.at("08:00").do(etl_process)

# Run the scheduled tasks
while True:
    schedule.run_pending()
    time.sleep(1)
```
