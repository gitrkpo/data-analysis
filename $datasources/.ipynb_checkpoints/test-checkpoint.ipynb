```python
import pandas as pd 
data = pd.read_csv('dirtydata.csv') #read csv file

data['Calories']=data['Calories'].fillna(data['Calories'].mean()) #Take mean of Blanks

data['Date']=data['Date'].dropna() #drop na in Date
data.loc[data['Duration']>100,'Duration']=75 #Convert anything more than 100 to 75
data = data.dropna(subset=['Date']) #drop NaN in Date
data['Date']=pd.to_datetime(data['Date']) #convert to datetime
data['Duration']=data['Duration'].replace(450,58) #replace 450 wiht 58
 
data[data['Pulse']>100]  #filter Pulse>100
data[['Duration','Date','Pulse']] #specific columns 
```
