```python
#ability to replace empty cell using mean,median,mode
import pandas as pd
oo=pd.read_csv('dirtydata.csv')
x=oo["Calories"].mean()
oo["Calories"].fillna(x, inplace = True)
print(oo.to_string())
```

```python
#ability to replace empty cell using median , used during trends
import pandas as pd
oo=pd.read_csv('dirtydata.csv')
x=oo["Calories"].median()
oo["Calories"].fillna(x, inplace = True)
print(oo.to_string())
```

```python
#ability to replace empty cell using mode 
import pandas as pd
oo=pd.read_csv('dirtydata.csv')
x=oo["Calories"].mode()[0]
oo["Calories"].fillna(x, inplace = True)
print(oo.to_string())
```

```python
#for largerdata set: we will loop through all the values in duration
#if value is higher than 120 then set it to 120

import pandas as pd
sharad=pd.read_csv('dirtydata.csv')
for i in sharad.index:
    if sharad.loc[i, "Duration"] > 120:
        sharad.loc[i, "Duration"] = 120
print(sharad.to_string())
 
```

```python
#looping to remove value higher than 120
import pandas as pd
sharad=pd.read_csv('dirtydata.csv')
for i in sharad.index:
    if sharad.loc[i, "Duration"] > 120:
        sharad.drop(i, inplace=True)
print(sharad.to_string())
```

```python
import pandas as pd 
df = {"Make": ("Kia", "Hyundai", "Toyota", "Honda"),
      "Model": ("Sorento", "Santa Fe", "Highlander", "CR-V"),
      "Price": (20, 30, 30, 30)}
df = pd.DataFrame(df)

print(df)


df['Make']=df['Make'].str.upper()

df['Make']=df['Make'].replace('Honda','HONDA')
df['Car']=df['Make']+ " " + df['Model']

df['Markup']=df['Price']*100

new_row = pd.DataFrame({"Car": ['Range Rover'], "Price": [32], "Markup": [3200]})
df = pd.concat([df, new_row], ignore_index=True)

print(df[["Car",'Price','Markup']])
```
